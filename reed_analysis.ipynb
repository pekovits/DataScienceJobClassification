# Tfidf without normal stopwords

```python
##	This function implements the tfidf model for the corpus of job descriptions
#	@param the job descriptions
#	@return a tuple with the tfidf matrix, the vectorizer and the array with the values
def tfidf_model(descriptions):
	print "Creating the tf-idf model...\n"

	# intializing the TfidfVectorizer from scikit-learn
	# TfidfVectorizer is a class that combines all the options 
	# of CountVectorizer TfidfTransformer in a single model
	vectorizer = TfidfVectorizer(max_df=0.8, max_features=200000,
                                 min_df=0.2, stopwords = 'english',
                                 use_idf=True, ngram_range=(1,3))

	# fit the vectorizer to the descriptions 
	# learns the vocabulary and idf, and returns a term document matrix
	tfidf_matrix = vectorizer.fit_transform(descriptions)

	# transform the matrix to a numpy array
	# term_array = tfidf_matrix.toarray()
	term_array = np.asarray(tfidf_matrix)

	# vocabulary_lookup(vectorizer, term_array)

	return (tfidf_matrix, vectorizer, term_array)
```

```python
x, d, f = tfidf_model(description_no_stopwords)
freqs = [(word, x.getcol(idx).sum()) for word, idx in d.vocabulary_.items()]
for f in sorted (freqs, key = lambda x: -x[1]):
    print f
```
